```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

### Used below features in x1 variable

```python
bestfeatures=SelectKBest(score_func=chi2,k=5)
fit=bestfeatures.fit(x,y)
# calculate the scores
dfscore=pd.DataFrame(fit.scores_)
# get column names into variable
dfcolumns=pd.DataFrame(x.columns)
# concat 2 dataframe above
feat_col=pd.concat([dfcolumns,dfscore],axis=1)
feat_col.columns=['Specs','Score'] # Naming the DataFrame
print(feat_col.nlargest(7,'Score'))
```

### Used below features in x2 variable

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(x,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=x.columns)
feat_importances.nlargest(7).plot(kind='bar')
plt.show()
```

### Testing with different columns in independent variable

```python
dataset.info()
```

```python
x1=dataset.loc[:,['legs','feathers','fins','milk','airborne','hair','eggs']].values
x2=dataset.loc[:,['legs','feathers','fins','milk','backbone','hair','eggs','toothed']].values
x3=dataset.loc[:,['legs','feathers','fins','milk','backbone','domestic','eggs','toothed','venomous']].values
```

# Using Minkowski Method

## Training model on x1 feature set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x1, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score # Score of 80.95%
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Training model on x2 feature set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x2, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score # Score of 90.47%
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
